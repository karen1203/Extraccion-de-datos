```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# carga de un archivo .xlsx sin indice
df= pd.read_excel("Detalle precios y productos fabricados 2022.xlsx")
df
```

```python
#flitro por objeto
filtro1= df[df["NOMBRE_VENDEDOR"]=="LETICIA RAMIREZ HERNANDEZ"]
filtro1
```

```python
#CONVERTIR ARCHIVO A CSV
filtro1.to_csv("Entregable1.csv")
```

```python
#filtro por fila
filtro2=df.iloc[[3,12],:]
filtro2
```

```python
#CONVERTIR ARCHIVO A CSV
filtro2.to_csv("Entregable2.csv")
```

```python
#filtro por columna
filtro3=df.iloc[: ,[3,5]] #filas no consecutivas
filtro3
```

```python
#CONVERTIR ARCHIVO A CSV
filtro3.to_csv("Entregable3.csv")
```

```python
df1=pd.read_excel("Detalle precios y productos fabricados 2022.xlsx", index_col=2)
df1
```

```python
#filtro filas vs columnas
filtro4= df1.loc[["AARON NUÑO CORDOVA","LETICIA RAMIREZ HERNANDEZ"],["NOMBRE_CLIENTE"]]
filtro4
```

```python
#CONVERTIR ARCHIVO A CSV
filtro4.to_csv("Entregable4.csv")
```

```python
#filtro por cabecera
filtro5=df.head(10)
filtro5
```

```python
#CONVERTIR ARCHIVO A CSV
filtro5.to_csv("Entregable5.csv")
```

```python
#filtro de cola
filtro6=df.tail(10)
filtro6
```

```python
#CONVERTIR ARCHIVO A CSV
filtro6.to_csv("Entregable6.csv")
```

```python
#Filtro por comparación
filtro7=df[df["PRECIO_UNITARIO"]>7000]
filtro7
```

```python
#CONVERTIR ARCHIVO A CSV
filtro7.to_csv("Entregable7.csv")
```

```python
#filtro Y
filtro8=df[(df["PRECIO_UNITARIO"]<6000)&(df["FECHA_DOC"]=="2022-03-14")]
filtro8
```

```python
#CONVERTIR ARCHIVO A CSV
filtro8.to_csv("Entregable8.csv")
```

```python
#filtro o
filtro9=df[(df["PRECIO_UNITARIO"]<6000)| (df["FECHA_DOC"]=="2022-03-14")]
filtro9
```

```python
#CONVERTIR ARCHIVO A CSV
filtro9.to_csv("Entregable9.csv")
```

```python
#filtro not
filtro10=df[(df["PRECIO_UNITARIO"]<6000)&~(df["FECHA_DOC"]=="2022-03-14")]
filtro10
```

```python
#CONVERTIR ARCHIVO A CSV
filtro10.to_csv("Entregable10.csv")
```
